```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
b_weiner=[62,43,67,39,15,32]
b_clique_number=[4,4,4,3,3,4]
b_number_cliques=[8,9,10,5,3,5]
b_random1=[4.249595633483295,2.7053656638747725,0,3.022284339659236,\
           2.6411864761932917,3.8153573673902006]
b_random2=[-0.07926169269759559,-8.179198646450617,0,-0.17228274273693464,\
           -0.2292275768018899,-0.059705991652010015]
```

```python
weiner=[2831,6258,4056,1839,5986]
clique_number=[2,2,2,2,2]
number_cliques=[33,9,41,30,45]
random1=[2.7053656638747725,3.8527789925489313,0.07407996950711104,\
         1.9539925233402755e-14,-3.57267939429139]
random2=[-8.179198646450617,-10.806815812439472,-11.34078411885495,\
         -10.206207261596562,-14.137478055716889]
```

```python
ninels=[35929,19780,21692,40543,43191,31080]
fivels=[34764,34287,30159,54541,37296,40624]
```

```python
x=b_weiner
y1=ninels
y2=fivels
plt.figure(figsize=(8,8))
plt.scatter(x,y2,color='blue',label='5x5')
plt.scatter(x,y1,color='red',label='9x9')
z = np.polyfit(x,y1, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")
z = np.polyfit(x,y2, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"b--")
plt.xlabel('Weiner Index')
plt.ylabel('Number of Steps taken to explore the map')
plt.title('Complexity profile for 9x9 and 5x5 exploration window')
plt.legend()
```
