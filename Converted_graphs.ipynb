```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G9a = nx.Graph()
for i in range(1,7):
    G9a.add_node(i)

#The original Graph
G9a.add_edge(1,2)
G9a.add_edge(1,3)
G9a.add_edge(1,6)
G9a.add_edge(2,3)
G9a.add_edge(2,5)
G9a.add_edge(3,4)
G9a.add_edge(3,6)
G9a.add_edge(4,6)
G9a.add_edge(5,5)


pos = nx.spring_layout(G9a)
nx.draw(G9a, pos, with_labels=True)
plt.show()
```

```python
#Converted corresponding line graph
G9b=nx.line_graph(G9a)
pos = nx.spring_layout(G9b)
nx.draw(G9b, pos, with_labels=True)
plt.show()
```

```python
#documentation https://networkx.org/documentation/stable/reference/algorithms/clique.html
nx.algorithms.clique.graph_clique_number(G9b)
```

```python
nx.algorithms.clique.graph_number_of_cliques(G9b)
```

```python
#documentation https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.non_randomness.non_randomness.html
nx.non_randomness(G9b)
```

```python
#documentation: https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.wiener.wiener_index.html
nx.algorithms.wiener.wiener_index(G9b)
```
